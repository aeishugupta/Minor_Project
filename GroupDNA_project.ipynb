```python
import numpy as np

with open('/content/hostel_bois.txt','r',encoding='utf-8') as f:
  lines = f.readlines()

  print("Total Lines:",len(lines))
  print()
  print(lines[0])
```

```python
for i in range(10):
  print(lines[i])
```

```python
line = lines[10]
print(line)
```

```python
parts = line.split(" - ")
print(parts)
```

```python
date_time = parts[0]
sender_message = parts[1]

date = date_time.split(", ")[0]
time = date_time.split(", ")[1]

sender = sender_message.split(": ",1)[0]
message = sender_message.split(": ",1)[1]

print("Date = ",date)
print("Timme = ",time)
print("Sender = ",sender)
print("Message = ",message)
```

```python
message_count = {}
for line in lines:
  if " - " not in line:
    continue
  parts = line.split(" - ")

  if len(parts) < 2:
    continue

  sender_message = parts[1]

  if ": " not in sender_message:
    continue

  sender, message = sender_message.split(": ",1)
  if sender in message_count:
    message_count[sender] += 1
  else:
    message_count[sender] = 1

print(message_count)
```

```python
message_values = np.array(list(message_count.values()))

print("Average messages =",np.mean(message_values))
print("Maximum messages =",np.max(message_values))
print("Minimum messages =",np.min(message_values))
```

```python
# To find most active user

most_active = ""
for person in message_count:
  if most_active == "":
    most_active = person
  elif message_count[person] > message_count[most_active]:
    most_active = person

print("Most active person : ",most_active)
print("Total Messages = ",message_count[most_active])


# To find least active user

least_active = ""
for person in message_count:
  if least_active == "":
    least_active = person
  elif message_count[person] < message_count[least_active]:
    least_active = person

print("Least active person : ",least_active)
print("Total Messages = ",message_count[least_active])
```

```python
# to count media messages
media_count = 0
for line in lines:
  if "<Media omitted>" in line:
    media_count += 1
print("Total Media Messages = ",media_count)
```

```python
# To count deleted messages

deleted_count = 0
for line in lines:
  if "This message was deleted" in line:
    deleted_count += 1
print("Total Deleted messages = ",deleted_count)
```

```python
night_count = {}

for line in lines:

  if " - " not in line:
    continue

  parts = line.split(" - ")
  if len(parts) < 2:
    continue

  date_time = parts[0]
  sender_message = parts[1]
  if ": " not in sender_message:
    continue

  sender = sender_message.split(": ",1)[0]
  time = date_time.split(", ")[1]
  hour = int(time.split(":")[0])
  if hour >= 23 or hour <= 4:
    if sender in night_count:
      night_count[sender] += 1
    else:
      night_count[sender] = 1
print(night_count)
```

```python
night_owl = ""
for person in night_count:
  if night_owl == "":
    night_owl = person
  elif night_count[person] > night_count[night_owl]:
    night_owl = person
print("Night Owl is :",night_owl)
print("Night Messages :",night_count[night_owl])
```

```python
# Most Used Words
word_count = {}

for line in lines:

  if " - " not in line:
    continue
  parts = line.split(" - ")
  if len(parts) < 2:
    continue
  sender_message = parts[1]
  if ": " not in sender_message:
    continue

  message = sender_message.split(":",1)[1]
  words = message.split()

  for word in words:
    word = word.lower()
    if word in word_count:
      word_count[word] += 1
    else:
      word_count[word] = 1
print(word_count)

```

```python
# To find most used word

most_word = ""
for word in word_count:
  if most_word == "":
    most_word = word
  elif word_count[word] > word_count[most_word]:
    most_word = word
print("Most Used Word :",most_word)
print("Count :",word_count[most_word])
```

```python
# To print top 10 words

temp = word_count.copy()
print("TOP 10 WORDS: ")
for i in range(10):
  max_word = ""
  for word in temp:
    if max_word == "":
      max_word = word
    elif temp[word] > temp[max_word]:
      max_word = word
  print(max_word, "=",temp[max_word])
  del temp[max_word]
```

```python
# To count messages in each hour

hour_count = {}
for line in lines:
  if " - " not in line:
    continue
  parts = line.split(" - ")
  if len(parts) < 2:
    continue
  date_time = parts[0]
  time = date_time.split(", ")[1]
  hour = time.split(":")[0]
  if hour in hour_count:
    hour_count[hour] += 1
  else:
    hour_count[hour] = 1
print(hour_count)
```

```python
busy_hour = ""
for hour in hour_count:
  if busy_hour == "":
    busy_hour = hour
  elif hour_count[hour] > hour_count[busy_hour]:
    busy_hour = hour
print("Busiest hour :",busy_hour)
print("Messages :",hour_count[busy_hour])
```

```python
day_count = {}

for line in lines:

  if " - " not in line:
    continue

  parts = line.split(" - ")

  if len(parts) < 2:
    continue

  date_time = parts[0]

  date = date_time.split(",")[0]

  if date in day_count:
    day_count[date] += 1
  else:
    day_count[date] = 1

print(day_count)
```

```python
busy_day = ""

for day in day_count:
  if busy_day == "":
    busy_day = day
  elif day_count[day] > day_count[busy_day]:
    busy_day = day
print("Busiest day :",busy_day)
print("Messages :",day_count[busy_day])
```

```python
least_day = ""

for day in day_count:
  if least_day == "":
    least_day = day
  elif day_count[day] < day_count[least_day]:
    least_day = day
print("Least Active Day :",least_day)
print("Messages :",day_count[least_day])
```
